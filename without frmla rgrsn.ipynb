```python

import numpy as np
from sklearn.linear_model import LinearRegression

Area = np.array([504,743,752,813,906,1206,1446,1532,1606,1679,1906,1956,2106,3778]).reshape((-1,1))
Price = np.array([22.1,29.85,33,35.22,38,36,58.56,51.5,51,76,73,71,108,131])

model = LinearRegression().fit(Area,Price)
print('slope , m :', model.coef_)
print('Intercept ,c :', model.intercept_)
```

```python
#calculate price for 1000 sq.ft
Area_new = np.array([1000]).reshape((-1, 1))
print("Built-up Area :",Area_new)
y_new = model.predict(Area_new)
print("built up area price for 1000 sq.ft : ", y_new , "in lakhs")
```

```python
#ploting line and scatter plot
import matplotlib.pyplot as plt
max_x = np.max(Area) + 100
min_x = np.min(Price) - 100

rx = np.linspace(min_x, max_x, 1000)
ry = model.intercept_ + model.coef_ * rx 
 
plt.plot(rx, ry, color='#9198b8', label='Regression Line')

plt.scatter(Area, Price, c='#8fff82', label='Scatter Plot')
 
plt.xlabel('Area')
plt.ylabel('Price')
plt.legend()
plt.show()
```
